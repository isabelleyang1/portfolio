# Feature Engineering

```python
cleaned_df["SalesLineTotal"] = cleaned_df["Quantity"] * cleaned_df["Price"]
```

```python
cleaned_df
```

```python
aggregated_df = cleaned_df.groupby(by="Customer ID", as_index=False) \
    .agg(
        MonetaryValue=("SalesLineTotal", "sum"), #total spent by customer
        Frequency=("Invoice", "nunique"), # no. of transactions
        LastInvoiceDate=("InvoiceDate", "max") #recency of transaction
    )
aggregated_df.head(5)
```

```python
max_invoice_date = aggregated_df["LastInvoiceDate"].max()

aggregated_df["Recency"] = (max_invoice_date - aggregated_df["LastInvoiceDate"]).dt.days

aggregated_df.head(5)
```

```python
M_Q1 = aggregated_df['MonetaryValue'].quantile(0.25)
M_Q1
```

```python
M_Q3 = aggregated_df['MonetaryValue'].quantile(0.75)
M_Q3
```

```python
M_IQR = M_Q3 - M_Q1

monetary_outliers_df = aggregated_df[(aggregated_df['MonetaryValue'] > (M_Q3 + 1.5 * M_IQR)) \
| (aggregated_df['MonetaryValue'] < (M_Q1 - 1.5 * M_IQR))].copy()

monetary_outliers_df.describe()
```

```python
F_Q1 = aggregated_df['Frequency'].quantile(0.25)
F_Q3 = aggregated_df['Frequency'].quantile(0.75)
F_IQR = F_Q3 - F_Q1

frequency_outliers_df = aggregated_df[(aggregated_df['Frequency'] > (F_Q3 + 1.5 * F_IQR)) \
| (aggregated_df['Frequency'] < (F_Q1 - 1.5 * F_IQR))].copy()

frequency_outliers_df.describe()
```

```python
non_outliers_df = aggregated_df[(~aggregated_df.index.isin(monetary_outliers_df.index)) & (~aggregated_df.index.isin(frequency_outliers_df.index))]

non_outliers_df.describe()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1) #1 row, 3 columns, 1st graph
sns.boxplot(data=non_outliers_df['MonetaryValue'], color='skyblue')
plt.title('Monetary Value Boxplot')
plt.xlabel('Monetary Value')

plt.subplot(1, 3, 2) #1 row, 3 columns, 2nd graph
sns.boxplot(data=non_outliers_df['Frequency'], color='lightgreen')
plt.title('Frequency Boxplot')
plt.xlabel('Frequency')

plt.subplot(1, 3, 3) #1 row, 3 columns, 3rd graph
sns.boxplot(data=non_outliers_df['Recency'], color='salmon')
plt.title('Recency Boxplot')
plt.xlabel('Recency')

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))

ax = fig.add_subplot(projection="3d")

scatter = ax.scatter(non_outliers_df["MonetaryValue"], non_outliers_df["Frequency"], non_outliers_df["Recency"])

ax.set_xlabel("Monetary Value")
ax.set_ylabel("Frequency")
ax.set_zlabel("Recency")

ax.set_title('3d Scatter Plot of Customer Data')

plt.show()
```

```python
scaled_data_df = pd.DataFrame(scaled_data, index=non_outliers_df.index, columns=("MonetaryValue", "Frequency", "Recency"))

scaled_data_df
```

# KMeans Clustering

```python
max_k = 12

inertia = []
silhoutte_scores = []
k_values = range(2, max_k + 1)

for k in k_values:

    kmeans = KMeans(n_clusters=k, random_state=42, max_iter=1000)

    cluster_labels = kmeans.fit_predict(scaled_data_df)

    sil_score = silhouette_score(scaled_data_df, cluster_labels)

    silhoutte_scores.append(sil_score)

    inertia.append(kmeans.inertia_)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(k_values, inertia, marker='o')
plt.title('KMeans Inertia for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(k_values, silhoutte_scores, marker='o', color='orange')
plt.title('Silhouette Scores for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_values)
plt.grid(True)

plt.tight_layout()
plt.show()
```

silhouette score - measures overlap between clusters
4 clusters has less potential overlap than 5 cluster option

```python
kmeans = KMeans(n_clusters=4, random_state=42, max_iter=1000)

cluster_labels = kmeans.fit_predict(scaled_data_df)

cluster_labels
```

```python
non_outliers_df["Cluster"] = cluster_labels

non_outliers_df
```

```python
cluster_colors = {0: '#1f77b4',  # Blue
                  1: '#ff7f0e',  # Orange
                  2: '#2ca02c',  # Green
                  3: '#d62728'}  # Red

colors = non_outliers_df['Cluster'].map(cluster_colors)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

scatter = ax.scatter(non_outliers_df['MonetaryValue'], 
                     non_outliers_df['Frequency'], 
                     non_outliers_df['Recency'], 
                     c=colors,  # Use mapped solid colors
                     marker='o')

ax.set_xlabel('Monetary Value')
ax.set_ylabel('Frequency')
ax.set_zlabel('Recency')

ax.set_title('3D Scatter Plot of Customer Data by Cluster')

plt.show()
```

```python
plt.figure(figsize=(12, 18))

plt.subplot(3, 1, 1)
# Plot the gray baseline violin first
sns.violinplot(x=[""] * len(non_outliers_df), y=non_outliers_df['MonetaryValue'], color='gray', linewidth=1.0, inner=None)
# Overlay with cluster-specific violins
sns.violinplot(x=non_outliers_df['Cluster'], y=non_outliers_df['MonetaryValue'], palette=cluster_colors, hue=non_outliers_df["Cluster"])
plt.title('Monetary Value by Cluster')
plt.ylabel('Monetary Value')

plt.subplot(3, 1, 2)
# Gray baseline violin
sns.violinplot(x=[""] * len(non_outliers_df), y=non_outliers_df['Frequency'], color='gray', linewidth=1.0, inner=None)
# Cluster-specific violins
sns.violinplot(x=non_outliers_df['Cluster'], y=non_outliers_df['Frequency'], palette=cluster_colors, hue=non_outliers_df["Cluster"])
plt.title('Frequency by Cluster')
plt.ylabel('Frequency')

plt.subplot(3, 1, 3)
# Gray baseline violin
sns.violinplot(x=[""] * len(non_outliers_df), y=non_outliers_df['Recency'], color='gray', linewidth=1.0, inner=None)
# Cluster-specific violins
sns.violinplot(x=non_outliers_df['Cluster'], y=non_outliers_df['Recency'], palette=cluster_colors, hue=non_outliers_df["Cluster"])
plt.title('Recency by Cluster')
plt.ylabel('Recency')

plt.tight_layout()
plt.show()
```

1. Cluster 0 (Blue): New Customers ENGAGE
- Rationale: This cluster represents the customers with the lowest monetary value, low purchase frequency, but high recency
- Action: Encourage higher spending and more frequent visits with introductory offers or loyalty incentives.
2. Cluster 1 (Orange): Loyal Customers REWARD
- Rationale: This cluster represents the customers with the highest monetary value, the highest purchase frequency, and low recency.
- Action: Use re-engagement campaigns like personalized messages or exclusive discounts to bring them back and maintain their loyalty.
3. Cluster 2 (Green): Consistent but Moderate Spenders NURTURE
- Rationale: This cluster represents the customers with moderate monetary value, moderate purchase frequency, and moderately high recency.
- Action: Nurture these customers with promotions or incentives that encourage them to increase their spending/frequency.
4. Cluster 3 (Red): Dormant/Lapsed Customers RE-ENGAGE
- Rationale: This cluster represents the custeomrs with moderate monetary value, purchase frequency, and recency.
- Action: Try win-back campaigns to re-engage them, i.e. "We Miss You" offers or limited-time discounts to rekindle their interest

```python
overlap_indices = monetary_outliers_df.index.intersection(frequency_outliers_df.index)

monetary_only_outliers = monetary_outliers_df.drop(overlap_indices)
frequency_only_outliers = monetary_outliers_df.drop(overlap_indices)
monetary_and_frequency_only_outliers = monetary_outliers_df.loc[overlap_indices]
```

```python
monetary_only_outliers["Cluster"] = -1
frequency_only_outliers["Cluster"] = -2
monetary_and_frequency_only_outliers["Cluster"] = -3

outlier_clusters_df = pd.concat([monetary_only_outliers, frequency_only_outliers, monetary_and_frequency_only_outliers])

outlier_clusters_df
```

```python
cluster_colors = {-1: '#9467bd',
                  -2: '#8c564b',
                  -3: '#e377c2'}

plt.figure(figsize=(12, 18))

plt.subplot(3, 1, 1)
# Plot the gray baseline violin first
sns.violinplot(x=[""] * len(outlier_clusters_df), y=outlier_clusters_df['MonetaryValue'], color='gray', linewidth=1.0, inner=None)
# Overlay with cluster-specific violins
sns.violinplot(x=outlier_clusters_df['Cluster'], y=outlier_clusters_df['MonetaryValue'], palette=cluster_colors, hue=outlier_clusters_df["Cluster"])
plt.title('Monetary Value by Cluster')
plt.ylabel('Monetary Value')

plt.subplot(3, 1, 2)
# Gray baseline violin
sns.violinplot(x=[""] * len(outlier_clusters_df), y=outlier_clusters_df['Frequency'], color='gray', linewidth=1.0, inner=None)
# Cluster-specific violins
sns.violinplot(x=outlier_clusters_df['Cluster'], y=outlier_clusters_df['Frequency'], palette=cluster_colors, hue=outlier_clusters_df["Cluster"])
plt.title('Frequency by Cluster')
plt.ylabel('Frequency')

plt.subplot(3, 1, 3)
# Gray baseline violin
sns.violinplot(x=[""] * len(outlier_clusters_df), y=outlier_clusters_df['Recency'], color='gray', linewidth=1.0, inner=None)
# Cluster-specific violins
sns.violinplot(x=outlier_clusters_df['Cluster'], y=outlier_clusters_df['Recency'], palette=cluster_colors, hue=outlier_clusters_df["Cluster"])
plt.title('Recency by Cluster')
plt.ylabel('Recency')

plt.tight_layout()
plt.show()
```

Cluster -1 (Monetary Outliers) PAMPER
- Characteristics: High spenders but not necessarily frequent buyers
- Action: Focus on maintaining their loyalty with personalized offers or luxury services that cater to their spending capacity

Cluster -2 (Frequency Outliers) UPSELL
- Characteristics: low spending, high frequency
- Action: Engage them with loyalty programs or bundle deals to encourage higher spending per visit

Cluster -3 (Monetary and Frequency Outliers) DELIGHT
- Characteristics: extreme spending and frequent purchases
- Action: develop VIP programs or exclusive offers to maintain their loyalty and encourage continued engagement

```python
cluster_labels = {
    0: "ENGAGE",
    1: "REWARD",
    2: "NURTURE",
    3: "RE-ENGAGE",
    -1: "PAMPER",
    -2: "UPSELL",
    -3: "DELIGHT"
}
```

```python
full_clustering_df = pd.concat([non_outliers_df, outlier_clusters_df])

full_clustering_df
```

```python
full_clustering_df["ClusterLabel"] = full_clustering_df["Cluster"].map(cluster_labels)

full_clustering_df
```
